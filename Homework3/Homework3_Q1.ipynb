```
#Ans:1
from numpy import *
import matplotlib.pyplot as plt
# Let P=[R, F] where R= number of rabits and F= number of foxes 
R0=100
F0=100
P0=array([R0,F0]) #initial population
Beta_r = 0.01
Beta_f = 0.005
del_r = 0.005
del_f = 0.01
ep_r = 0.0001
ep_f = 0.0001

# Defining Fox and Rabit population change with time from the slide equation
def dP_dt(P, t=0):
    return array([ Beta_r*P[0] - del_r*P[0] - ep_f*P[0]*P[1] ,
                  Beta_f*P[1] - del_f*P[1] + ep_r*P[0]*P[1] ])
```

```
from scipy.integrate import odeint
t = linspace(0, 10000,  1000)         # time
P= odeint(dP_dt, P0, t)                 # Population
R, F = P.T
```

```
#Plotting the population growth
p = plt.figure(figsize=(12,8))
plt.plot(t, R, 'g-', label='Rabbits')
plt.plot(t, F  , 'r-', label='Foxes')
plt.legend(loc='best')
plt.xlabel('time (days)')
plt.ylabel('population')
plt.title('Evolution of population')
```
