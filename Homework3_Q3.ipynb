```
import numpy as np
import matplotlib.pyplot as plt
```

```

r_235 = np.array([4321,4303,4153,3155,3672,2808,4355,4307,4078,4287,3651,3317,4392,3985])
r_238 = np.array([4233,4201,3780,1713,2595,1177,4356,4235,3674,4126,2604,1961,4364,3355])

ratio_235 = r_235*(10**6)
ratio_238 = r_238*(10**6)


ts = np.arange(0,6000,100)
t = ts *(10**6)
lambda_235 = 9.85 * (10**(-10)) #lambda=.693/half-life
lambda_238 = 1.54 * (10**(-10))

Cratio_235 = (np.exp(lambda_235*t)) - 1
Cratio_238 = (np.exp(lambda_238*t)) - 1
Dratio_235 =(np.exp(lambda_235*ratio_235)) - 1
Dratio_238 =(np.exp(lambda_238*ratio_238)) - 1

```

```

z = np.polyfit(Dratio_235,Dratio_238, 1)
p = np.poly1d(z)

xp = np.linspace(-1, 120, 1000)

f1 = plt.figure(figsize=(10,10))

plt.plot(Cratio_235 , Cratio_238, 'b-',label = 'Concordia')
plt.plot(Dratio_235 , Dratio_238, 'go', label = 'Discordia Data')
_=plt.plot(Dratio_235 , Dratio_238,'.', xp, p(xp), 'r-')

plt.grid()
plt.legend(loc='best')
plt.ylabel('206Pb/238U')
plt.xlabel('207Pb/235U')
plt.title('Concordia and discordia')
```
