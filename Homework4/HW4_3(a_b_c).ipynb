```
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
#3(a)
M=10**(-5)*np.array([5.08, 3.87, 3.02, 2.10, 1.18])
D=np.array([60.8, 62.1, 61.2, 62.2, 63.1])
I=np.array([60.1, 59.8, 62.1, 61.6, 60.9])
```

```
N=M*np.cos(np.radians(I))*np.cos(np.radians(D)) #calculating N-component
print (N)
#The N-component values (A/m) for temperature 100,200,300,400,500 and 550 respectively 
```

```
E=M*np.cos(np.radians(I))*np.sin(np.radians(D))  #calculating E-component
print (E)
#The E-component values (A/m) for temperature 100,200,300,400,500 and 550 respectively 
```

```
V=M*np.sin(np.radians(I))  #calculating V-component
print (V)
#The V-component values (A/m) for temperature 100,200,300,400,500 and 550 respectively 
```

```
#3(b)
#plotting E-component vs N-component
p1=np.polyfit(N,E,1)
pl1=np.poly1d(p1)
plt.figure(1,figsize=(8,8))
sp1=np.linspace(0,1e-6*15,100)
plt.xlabel('N-component (A/m)')
plt.ylabel('E-component (A/m)')
plt.plot(N,E,'or',sp1,pl1(sp1))
print(pl1)
```

```
#3(c)
#plotting N-component vs V-component
p2=np.polyfit(V,N,1)
pl2=np.poly1d(p2)
plt.figure(1,figsize=(8,8))
sp2=np.linspace(0,1e-6*50,100)
plt.xlabel('V-component (A/m)')
plt.ylabel('N-component (A/m)')
plt.plot(V,N,'or',sp2,pl2(sp2))
print(pl2)
```

```
#plotting E component vs V-component
p3=np.polyfit(V,E,1)
pl3=np.poly1d(p3)
plt.figure(1,figsize=(8,8))
sp3=np.linspace(0,1e-6*50,100)
plt.xlabel('V-component (A/m)')
plt.ylabel('E-component (A/m)')
plt.plot(V,E,'or',sp3,pl3(sp3))
print(pl3)
```
